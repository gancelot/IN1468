# 17_grouping
### Demonstrates the use of Pandas groupby() feature.

```python
import pandas as pd
```

```python
contacts = pd.read_csv('contacts2.dat', header=None,
                       names=['name', 'address', 'state', 'zip', 'area_code', 'phone', 'email', 'company', 'position'],
                       converters={'state': lambda txt: txt.strip()})
```

```python
contacts
```

```python
bystate = contacts.groupby('state')
```

```python
bystate.groups
```

```python
bystate.nth(1)
```

```python
bystate.nth(1)['name']
```

```python
bystate.size()
```

```python
bystate.indices
```

```python
bystate.first()
```

```python
bystate.get_group('CO')
```

```python
colorado = bystate.get_group('CO')

colorado[colorado['zip'] == 82101]
```
