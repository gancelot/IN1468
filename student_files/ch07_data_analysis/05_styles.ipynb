### Use of style attributes within Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_ticks = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
x = np.arange(1, 13)
y = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
```

```python
plt.xlabel('Month')
plt.ylabel('Number of Days')
plt.xticks(x, x_ticks)
plt.plot(x, y, 'c--');
```

```python
plt.plot(x, y, 'rs:');
```

```python
plt.plot(x, y, 'mx-');
```

```python
plt.plot(x, y, 'g*-.');
```
